```python
from termcolor import colored
from random import choices as prob_rand
from random import shuffle

def outm(map):
  for x in map:
    for y in x:
      print(tiles[y],end='')
      #print(tiles[y],end='')
    print('')
```

```python
letter = 'â–ˆâ–ˆ'
tiles = [colored(letter, 'blue'),colored(letter, 'yellow'),colored(letter, 'green'), colored(letter,'white'),colored(letter,'red')]
#tiles = ['ðŸŸ¦','ðŸŸ¨','ðŸŸ©', colored(letter,'white'), 'ðŸŸ«']

tiles_num = [0,1,2,3,4]
```

```python
def gen_map(N = 50, M = 100, layers = 2, land = 0.3):

  lis = []
  for x in range(1,N-1):
    for y in range(1,M-1):
      lis+=[[x,y]]
  shuffle(lis)

  map = []
  for i in range(N):
    map+=[[3 for j in range(M)]]
  for i in range(N):
    for j in range(M):
      if((i%(N-1))*(j%(M-1))==0):       #kolorowanie granic
        map[i][j] = 0

  emcnt = 0
  for _k in range(layers):
    for xy in lis:
      x = xy[0]
      y = xy[1]
      cnt = [0,0,0,0,0]
      cnt[map[x+1][y]] += 1
      cnt[map[x-1][y]] += 1
      cnt[map[x][y-1]] += 1
      cnt[map[x][y+1]] += 1

  # 0 -> water
  # 1 -> sand
  # 2 -> land

      #nothing
      if(cnt[3] == 4):
        map[x][y] = prob_rand(tiles_num, weights = (100*(1-land),0,100*land,0,0) )[0]
        emcnt += 1

      #deep land -> deep land / land
      elif(cnt[4] > 0 and cnt[0] == 0):
        map[x][y] = prob_rand(tiles_num, weights = (0,0,10,0,50) )[0]

      #water + land -> sand
      elif(cnt[0] > 0 and cnt[2]+cnt[4] > 0):
        map[x][y] = prob_rand(tiles_num, weights = (0,1,0,0,0) )[0]

      #water -> water / sand
      elif(cnt[0] > 0):
        map[x][y] = prob_rand(tiles_num, weights = (100,1,0,0,0) )[0]

      #land -> land
      elif(cnt[2]+cnt[4] > 0 and cnt[1] == 0):
        map[x][y] = prob_rand(tiles_num, weights = (0,0,1,0,0) )[0]

      #land + sand -> land / sand
      elif(cnt[2]+cnt[4] > 0 and cnt[1] > 0):
        map[x][y] = prob_rand(tiles_num, weights = (0,1,100,0,0) )[0]

      else:
        map[x][y] = prob_rand(tiles_num, weights = (1,1,1,0,0) )[0]
    if(_k==1 and False):
      outm(map)
      print("\n")
  outm(map)
```

```python
gen_map(40,60,10,0.4)
```
